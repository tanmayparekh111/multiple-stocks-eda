```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stocklist=['TATASTEEL.NS','TATAMOTORS.NS','SBIN.NS','WIPRO.NS','ITC.NS']
stock=yf.download(stocklist,start='2017-06-10',end="2022-06-09")
```

```python
stock
```

```python
df=stock.Close.copy()
df
```

```python
df.columns = ['ITC', 'SBIN', 'TATAMOTORS', 'TATASTEEL', 'WIPRO']
```

```python
df
```

```python
%matplotlib inline
plt.style.use('seaborn')
df.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=13)
plt.show()
```

```python
df.iloc[0]
normalclose=df.div(df.iloc[0]).mul(100)
normalclose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=13)
plt.show()
```
